```python
def find_katz_fd(df):
  add_values=[]
  def katz_fd(series):
    n = np.abs(np.ediff1d(series)).sum()  # Sum of distances between consecutive points
    # print(n)
    L = len(series)  # Length of the signal
    # print(L)
    d = np.max(np.abs(series - series[0]))  # Estimated diameter
    # print(d)
    if d == 0 or L == 0:
      return 0  # Or any other value you find appropriate
    katz_fd_value = np.log10(n) / (np.log10(d / L) + np.log10(n))

    return katz_fd_value


  # Assuming 'EEG' is your DataFrame and 'FP_2' is the column you're interested in
  # Assuming df is your DataFrame
  mean_value = df['FP2'].mean()

  # Extract the 'FP_2' column as a numpy array
  fp_2_column = df['FP2'].values
  # print(fp_2_column)

  # Calculate Katz FD for 'FP_2' column
  katz_fd_value = katz_fd(fp_2_column)
  if katz_fd_value==0:
    add_values.append(mean_value)
  else:
    add_values.append(katz_fd_value)

  Katz_FD_Non_alcoholic.loc[i] = add_values
  # Print the result
  # print("Katz FD for 'FP_2' column:", katz_fd_value)
```
