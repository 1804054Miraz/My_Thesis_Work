```python
EEG1 = EEG_Features[['MAX', 'label']].copy()
X = EEG1.iloc[:,:-1]
y = EEG1.iloc[:,-1]



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Encode the target variable if it's categorical
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# K-Nearest Neighbors (KNN)
knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(X_train, y_train_encoded)
knn_predictions = knn_classifier.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test_encoded, knn_predictions))
# print("KNN Classification Report:")
# print(classification_report(y_test_encoded, knn_predictions))

# Support Vector Machine (SVM)
svm_classifier = make_pipeline(StandardScaler(), SVC())
svm_classifier.fit(X_train, y_train_encoded)
svm_predictions = svm_classifier.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test_encoded, svm_predictions))
# print("SVM Classification Report:")
# print(classification_report(y_test_encoded, svm_predictions))

# XGBoost
xgb_classifier = XGBClassifier()
xgb_classifier.fit(X_train, y_train_encoded)
xgb_predictions = xgb_classifier.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test_encoded, xgb_predictions))
# print("XGBoost Classification Report:")
# print(classification_report(y_test_encoded, xgb_predictions))

# Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train_encoded)
rf_predictions = rf_classifier.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test_encoded, rf_predictions))
# print("Random Forest Classification Report:")
# print(classification_report(y_test_encoded, rf_predictions))

# Convolutional Neural Network (CNN) - Example (not suitable for tabular data)
# ... code for CNN implementation ...

# Recurrent Neural Network (RNN) - Example (not suitable for tabular data)
# ... code for RNN implementation ...

# Multi-layer Perceptron (MLP) - Neural Network as an alternative
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp_classifier.fit(X_train, y_train_encoded)
mlp_predictions = mlp_classifier.predict(X_test)
print("MLP Accuracy:", accuracy_score(y_test_encoded, mlp_predictions))
# print("MLP Classification Report:")
# print(classification_report(y_test_encoded, mlp_predictions))
```
