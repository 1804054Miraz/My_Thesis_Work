```python
import pandas as pd
import matplotlib.pyplot as plt

# Read each CSV file into a dataframe
# delta_df = pd.read_csv('/content/delta_band.csv')
# theta_df = pd.read_csv('/content/theta_band.csv')
alpha_df = pd.read_csv('/content/drive/MyDrive/Dataset/EEG_FULL/Filter_FS_alcohol/alpha_ps_a_364_071_S2match.csv.csv')
# beta_df = pd.read_csv('/content/beta_band.csv')
# gamma_df = pd.read_csv('/content/gamma_band.csv')

# Function to plot the power spectrum for a given dataframe
def plot_power_spectrum(df, label_name):
    plt.figure(figsize=(12, 6))
    for column in df.columns[1:]:
        plt.plot(df['Frequency'], df[column], label=f'{label_name} - {column}')

    plt.yscale('log')  # Set y-axis to logarithmic scale
    plt.ylim(1e-12, 1e6)  # Adjust y-axis limits based on your data
    plt.title(f'Power Spectrum - {label_name}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Power/Frequency (uV^2/Hz)')
    # plt.legend()
    plt.show()

# Plot power spectrum for each band
plot_power_spectrum(alpha_df, 'Alpha Band')
# plot_power_spectrum(delta_df, 'Delta Band')
# plot_power_spectrum(theta_df, 'Theta Band')
# plot_power_spectrum(beta_df, 'Beta Band')
# plot_power_spectrum(gamma_df, 'Gamma Band')
```

```python
def find_mean(df, output_directory, out_directory, band, csv_file):

  output_file_path = os.path.join(output_directory, f'{band}_{csv_file}')
  df.to_csv(output_file_path, index=False)

  df=df.drop(columns=['Frequency'], axis=False)
  df = pd.DataFrame(df.mean()).transpose()
  band = band+'_mean_'
  out_file = os.path.join(out_directory, f'{band}_{csv_file}')
  df.to_csv(out_file, index=False)
```

```python
def export_folder(folder, input_directory):
  input_directory = os.path.join(input_directory, folder)
  path1=os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/Band_FS_Non-alcohol2/'+ folder)
  path2=os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/Features_PS_Non-alcohol2/'+ folder)
  os.mkdir(path1)
  os.mkdir(path2)
  time.sleep(3)

  output_directory = os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/Band_FS_Non-alcohol2/'+ folder+'/')
  out_directory = os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/Features_PS_Non-alcohol2/'+ folder+'/')

  # print(output_directory)
  # print(out_directory)
  # print(input_directory)
  # Define the frequency bands
  delta_band = (0, 4)
  theta_band = (4, 8)
  alpha_band = (8, 13)
  beta_band = (13, 30)
  gamma_band = (30, 100)

  filtered_df=pd.DataFrame()
  # Get a list of all CSV files in the directory
  csv_files = [file for file in os.listdir(input_directory) if file.endswith('.csv')]

  # Function to filter data into frequency bands and save to CSV
  def filter_and_save_band_data(df, band_name, band):
      filtered_df = df[(df['Frequency'] >= band_name[0]) & (df['Frequency'] <= band_name[1])]
      # print(filtered_df.head())
      return filtered_df
      # filtered_df.to_csv(os.path.join(output_directory, f'{band}_band.csv'), index=False)

  i=0
  alpha_='alpha'
  beta_='beta'
  delta_='delta'
  gamma_='gamma'
  theta_='theta'
  # Iterate over each CSV file in the directory
  for csv_file in csv_files:
    # print(csv_file)
    # Read the CSV file into a dataframe
    df = pd.read_csv(os.path.join(input_directory, csv_file))

    alpha_band_c=filter_and_save_band_data(df, alpha_band,'alpha')
    find_mean(alpha_band_c, output_directory, out_directory, alpha_, csv_file)

    beta_band_c=filter_and_save_band_data(df, beta_band,'beta')
    find_mean(beta_band_c, output_directory, out_directory, beta_, csv_file)

    delta_band_c = filter_and_save_band_data(df, delta_band, 'delta')
    find_mean(delta_band_c, output_directory, out_directory, delta_, csv_file)

    gamma_band_c=filter_and_save_band_data(df, gamma_band,'gamma')
    find_mean(gamma_band_c, output_directory, out_directory, gamma_, csv_file)

    theta_band_c=filter_and_save_band_data(df, theta_band,'theta')
    find_mean(theta_band_c, output_directory, out_directory, theta_, csv_file)

```

```python
import os
import pandas as pd
output_directory = '/content/drive/MyDrive/Dataset/Features/'
global cn
cn=0
# Specify the input and output directories
def directory_send(folder, input_directory):
  input_directory = os.path.join(input_directory,folder)
  csv_files = [file for file in os.listdir(input_directory) if file.endswith('.csv')]

  for csv_file in csv_files:
    global cn
    cn+=1
    global merged_alpha_df
    global merged_beta_df
    global merged_delta_df
    global merged_gamma_df
    global merged_theta_df
    # print(csv_file)
    df = pd.read_csv(os.path.join(input_directory, csv_file))

    if csv_file.startswith('alpha'):
      if not merged_alpha_df.empty:
        merged_alpha_df = pd.concat([merged_alpha_df, df], ignore_index=True)
      else:
        merged_alpha_df = df
    #   # print(merged_alpha_df)
    if csv_file.startswith('beta'):
      if not merged_beta_df.empty:
        merged_beta_df = pd.concat([merged_beta_df, df], ignore_index=True)
      else:
        merged_beta_df=df
    if csv_file.startswith('delta'):
      if not merged_delta_df.empty:
        merged_delta_df = pd.concat([merged_delta_df, df], ignore_index=True)
      else:
        merged_delta_df=df
    if csv_file.startswith('gamma'):
      if not merged_gamma_df.empty:
        merged_gamma_df = pd.concat([merged_gamma_df, df], ignore_index=True)
      else:
        merged_gamma_df=df
    if csv_file.startswith('theta'):
      if not merged_theta_df.empty:
        merged_theta_df = pd.concat([merged_theta_df, df], ignore_index=True)
      else:
        merged_theta_df=df
```
