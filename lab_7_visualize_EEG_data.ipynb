```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to visualize data with labeled frequency bands using Seaborn
def visualize_frequency_bands(data, color):
    # Plot the data for all frequency bands on a single plot using Seaborn
    sns.lineplot(data=data.mean(axis=1), color=color, linewidth=1.5)

# Set Seaborn style
sns.set(style="darkgrid")

# Visualize data for all frequency bands for Non-Alcoholic and Alcoholic data
plt.figure(figsize=(14, 8))

# Non-Alcoholic data (e.g., color it in orange)
non_alcoholic_files = os.listdir('/content/drive/MyDrive/Dataset/EEG_FULL/Transfer_new_non_alcoholic/337/')
for file in non_alcoholic_files:
    if file.endswith('.csv'):
        non_alcoholic_data = pd.read_csv(os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/Transfer_new_non_alcoholic/337/', file))
        visualize_frequency_bands(non_alcoholic_data, 'orange')

# Alcoholic data (e.g., color it in blue)
alcoholic_files = os.listdir('/content/drive/MyDrive/Dataset/EEG_FULL/New_alcoholic/364/')
for file in alcoholic_files:
    if file.endswith('.csv'):
        alcoholic_data = pd.read_csv(os.path.join('/content/drive/MyDrive/Dataset/EEG_FULL/New_alcoholic/364/', file))
        visualize_frequency_bands(alcoholic_data, 'blue')

# Add labels
plt.title('EEG Data - Alcoholic vs. Non-Alcoholic')
plt.xlabel('Time (Samples)')
plt.ylabel('Amplitude (μV)')

plt.show()
```
