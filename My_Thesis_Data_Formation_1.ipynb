```python
# df1.to_csv(f"{fn+'_000'}.csv")

# Assuming you have a list of DataFrames in dfs[]

# Initialize an empty array to store the CSV filenames
dcsv = []

# Specify the directory where you want to save the CSV files
output_directory = "/content/drive/MyDrive/Dataset/"

# Loop through the DataFrames and save them to CSV files
for i, df in enumerate(dfs):
    # Generate a filename for the CSV file (e.g., "file1.csv", "file2.csv", etc.)
    filename = f"file{i + 1}.csv"
    # Construct the full file path
    file_path = output_directory + filename
    # Save the DataFrame to a CSV file
    df.to_csv(file_path, index=False)  # Set index=False to avoid writing row numbers to the file
    # Append the filename to the dcsv array
    dcsv.append(filename)

# Print the list of CSV filenames
print("CSV filenames:", dcsv)
```
